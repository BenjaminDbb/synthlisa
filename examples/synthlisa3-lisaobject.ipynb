```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
import synthlisa
```

Demonstrating basic LISA object

```python
l = synthlisa.LISA(eccentric=True)
```

```python
ts = np.linspace(0,365.25 * 86400,100)
```

```python
xyz = np.fromiter((l.p(1,t) for t in ts),[('x','d'),('y','d'),('z','d')])
```

```python
for c in ['x','y','z']:
    pp.plot(xyz[c])
```

```python
l_c = synthlisa.LISA(eccentric=False)
```

```python
xyz_c = np.fromiter((l_c.p(1,t) for t in ts),[('x','d'),('y','d'),('z','d')])
```

```python
for c in ['x','y','z']:
    pp.plot(xyz[c] - xyz_c[c])
```
